# 고윳값의 개수
- N차방정식이 항상 N개의 복소수 해를 가진다는 사실을 이용하면 N차원 정방행렬의 고윳값의 개수에 대해 다음 정리가 성립힌다
    - 중복된 고윳값을 각각 별개로 생각하고 복수수인 고윳값도 고려한다면 N차원 정방핼영의 고윳값은 항상 N개다

# 고윳값과 대각합/행렬식
- 고윳값의 곱은 행렬식의 값과 같고 모든 고윳값의 합은 대각합의 값과 같다.

# 넘파이를 사용한고유분해

```python
A = np.array([[1, -2], [2, -3]])
w1, V1 = np.linalg.eig(A)

print(w1)
print(V1)
```

```python
B = np.array([[2, 3], [2, 1]])
w2, V2 = np.linalg.eig(B)

print(w2)
print(V2)
```

```python
C = np.array([[0, -1], [1, 0]])
w3, V3 = np.linalg.eig(C)

print(w3)
print(V3)
```

# 고윳값과 역행렬
- 대각화가능한 행렬에 0인 고유값이 없으면 항상 역행렬이 존재한ㄴ다

# 대칭행렬의 고유분해
- 행렬 A가 실수인 대칭행렬이면 고유값이 실수이고 고유벡터는 직교한다.
- 실수인 대칭행렬은 항상 대각화 가능하다.

# 대칭행렬을 랭크-1 행렬의 합으로 분해

# 대칭행렬의 고윳값 부호
- 대칭행렬이 양의 정부호이면 고윳값은 모두 양수이다. 역도 성립한다.
- 대칭행렬이 양의 준정부호이면 고윳값은 모두 0이거나 양수다. 역도 성립함다. 

# 분산행렬
- 분산행렬은 양의 준정부호이고 고윳값은 0보다 작거나 크다.

# 분산행렬의 역행렬
- 행렬 X가 풀랭크이면 이 행렬의 분산행렬 X.T X 의 역행렬이 존재한다. 

# N차원 정방행렬 A에 대해
- 행렬 A는 N개의 고윳값-고유벡터를 가진다.
- 행렬의 대각합은 모든 고윳값의 합과 같다.
- 행렬의 행렬식은 모든 고윳갑의 곱과 같다.
- 행렬 A가 대칭행렬이면 N개의 실수 고윳값을 가지며 고유벡터들이 서로 직교이다.
- 행렬 A가 대칭행렬이고 고윳값이 모두 양수이면 양의 정부호이고 역행렬이 존재한다. 역도 성립니다.
- 행려 A가 어떤 행렬 X의 분산행렬 X.T X이면 0또는 양의 고윳값을 가진다.
- 행렬 X가 풀랭크이면 분산행렬 X.T X은 역행렬이 존재한다.

# 특잇값 분해
- N X M 크기의 행렬A를 3개의 행렬의 곱으로 나타내는 것
- 특이분해는 모든 행렬에 대해 가능하다. 즉 어떤 행렬이 주어지더라도 특이분해할 수 있다. 

```python
from numpy.linalg import svd

A = np.array([[3, -1], [1, 3], [1, 1]])
U, S, VT = svd(A)
```

```python
U @ np.diag(S, 1)[:, 1:] @ VT
```
