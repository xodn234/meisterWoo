# 2.2 벡터와 행렬의 연산

- 모양이 같아야 연산이 가능하다.
- 브로드캐스팅(broadcasting)
- 평균제거(mean removed)벡터 혹은 0-평균(zero-mean)벡터
- 선형조합(linear combination) : 백터/행렬에 스칼라값을 곱한 후 더하거나 뺀 것
- 백터와 벡터의 곱셈(내적, 닷프로덕트) : 두 벡터의 길이는 같아야 한다 / 앞의 벡터가 행벡터 뒤의 벡터가 열 벡터야야 한다.
    - 덧셈은 벡터가되고, 곱셈은 스칼라가 된다
- 가중평균(weighted average) : 전체 가중치의 값의 합으로 나누면 가중평균이 된다.
- 유사도 : 두벡터가 닮은 정도를 정량적으로 나타낸 값 (비슷하면 값이 큼)

# 교환법칙과 분배법칙
- 행렬의 곱셈은 곱하는 행렬의 순서를 바꾸는 교환법칙이 성립하지 않는다. 그러나 덧셈에 대한 분배법칙은 성립한다.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = np.array([[9, 8], [7, 6]])
A @ B, B @ A, A @ (B + C), A @ B + A @ C, (A + B).T, A.T + B.T, (A @ B).T, B.T @ A.T
```

# 곱셈의 연결
- 연속된 행렬의 곱셈은 계산 순서를 임의의 순서로 해도 상관없다.
- 어떤 행렬이든 항등행렬을 곱하면 그 항렬의 값이 변하지 않는다.

```python
A = np.array([1, 2])
B = np.array([[1, 2], [3, 4]])
C = np.array([[5], [6]])
(A @ B) @ C, A @ (B @ C)
```

```python
A = np.array([[1, 2], [3, 4]])
I = np.eye(2)
A @ I, I @ A       
```
