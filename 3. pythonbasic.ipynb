```python
# 예제
# 계좌에 10000원이 들어 있습니다.
# 인출 금액을 입력 받습니다.
# 인출 금액이 계좌에 있는 금액보다 크면 인출이 불가능 합니다. 출력
# 인출 금액이 계좌에 있는 금액보다 작으면 "인출 되었습니다." 출력
# 마즈막에 현재 계좌의 잔액을 출력

data = int(input("인출 금액을 입력하세요 : "))
money = 10000

if data <= money:
    money -= data
    print(str(data) + "원이 인출되었습니다.")
else:
    print("인출이 불가능 합니다. " + str(data - money) + "원이 부족합니다.")
print("현재잔액은 : " + str(money) + "원 입니다.")
```

```python
# while

data = 3 # 조건이 False가 될때까지 굼누의 코드를 실행
while data:
    print(data)
    data -= 1
    
```

```python
# 예제
# 학생이 국어80, 영어90, 수학100점 입니다. while문을 이용하여 총점과 평균을 출력
# 학생의 점수는 list, dict 표현

subject = ["korean", "english", "math"]
point = [80, 90, 100]

total, avg = 0, 0

points = point.copy()

while points:
    total += points.pop()
avg = total / len(point)

total, avg
```

```python
# 무한루프
# break : 반복문을 중단 시킬떄 사용되는 예약어

result = 1
while result:
    if result >= 10:
        break
    result += 1
    
print(result)

```

```python
# for 
- iterable한 값을 하나씩 꺼내서 value에 대입시킨후 코드를 interable 변수의 값 갯수 만큼
  실행
    
for <variable> in <interables>:
    <code>
```

```python
# for - continue : 조건부분으로 올라가서 코드가 실행
ls = [0, 1, 2, 3, 4]
for data in ls: # data가 홀수가 되면 continue를 실행
    if data % 2:
        continue #data가 짝수이면 print를 실행
    print(data, end = " ")
```

```python
# args, kwargs
- 함수를 호출할때 아규먼트와 키워드 아규먼트의 갯수를 특정지을수 없을때 사용
```

```python
def plus(*args, **kwargs):
    print(type(args), args)
    print(type(kwargs), kwargs)
    return sum(args) + sum(kwargs.values())
```

```python
plus(1, 2, 3, num1=6, num2=7)
```

```python
### Scope 범위
- 함수 안에서 선언되는 변수와 함수 밖에서 선언되는 변수의 범위가 다릅니다. 
- global(전역), local(지역)
```

```python
# global

gv = 10

def echo():
    print(gv)
    
echo()
gv
```

```python
# local

gv = 10

def echo():
    gv = 100
    print(gv)
    
echo()
gv
```

```python
gv = 10

def echo():
    global gv
    gv = 100
    print(gv)
    
echo()
gv
```

```python
def minus (a, b):
    return a - b
```

```python
calc(plus, 1, 2)
```
