```python
# 클래스의 선언
class Calculator:
    
    num1 = 1
    num2 = 2
    
    def plus(self):
        return self.num1 + self.num2
    
    def minus(self):
        return self.num1 - self.num2
    
```

```python
# 클래스의 사용
calc = Calculator()
calc
```

```python
calc.num1, calc.num2, calc.plus(), calc.minus()
```

```python
# self의 의미 : 객체 자신
calc2 = Calculator()
```

```python
calc2.num1 = 10
```

```python
calc2.plus()
```

```python
## 생성자
- 클래스가 객체로 생성될때 실행되는 함수
- 변수(재료)를 추가할때 사용
```

```python
class Calculator:
    
    # 생성자 함수 : __init__
    def __init__(self, num1, num2=10):
        self.num1 = num1
        self.num2 = num2
    
    def plus(self):
        return self.num1 + self.num2
    
    def minus(self):
        return self.num1 - self.num2
```

```python
calc1 = Calculator(3, 4)
```

```python
calc1.plus()
```

```python
calc1 = Calculator(3)
```

```python
calc1.plus()
```

```python
# 클래스 문제
# 스타크래프트의 마린을 클래스로 설계
# 체력(health:40), 공격력(attack_pow:5), 공격(attack())
# 마린 클래스로 마린 객체 2개를 생성해서 마린 1이 마린 2를 공격하는 코드를 작성
# attack(self, unit)
```

```python
class Marine:
    def __init__(self, health=40, attack_pow=5):
        self.health = health
        self.attack_pow = attack_pow
        
    
    def attack(self, unit):
        unit.health -= self.attack_pow
        if unit.health <= 0:
            unit.health = 0
            print("사망")
        
        
```

```python
marine_1 = Marine()
```

```python
marine_2 = Marine()
```

```python
marine_1.attack(marine_2)
```

```python
marine_1.health, marine_2.health
```

```python
#  메딕 : heal_pow, heal(unit)

class Medic:
    def __init__(self, health=40, heal_pow=6):
        self.health = health
        self.heal_pow = heal_pow
        
    def heal(self, unit):
        if unit.health > 0:
            unit.health += self.heal_pow
            if unit.health >= 40:
                unit.health = 40
        else:
            print("이미 사망")
        
```

```python
medic = Medic()
```

```python
marine_1.attack(marine_2)
```

```python
marine_1.health, marine_2.health
```

```python
medic.heal(marine_2)
```
