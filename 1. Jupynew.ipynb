```python
### Jupiter Notebook
- mode
    - 명령모드 : 셀을 수정할때 사용
    - 편집모드 : 셀안의 내용을 수정할때 사용
- style
    - markdoun(명령모드 + m) : 셀안에 설명을 작성할때 사용
    - code (명령모드 + y): 파이썬 코드를 작성할때 사용
- 단축키
    - 셀 실행 : shift + enter
    - 셀 삭제 : 명령모드  x
    - 되돌리기 : 명령모드  z
    - 셀 생성 : 명령모드(위) a, (아래) b
```

```python
### Magic Command
-셀 내부에서 특별하게 동작하는 커멘드
- % : 한줄의 magic command를 동작
- %% : 셀단위의 magic command를 동작
- 주요 magic command
    - pwd 현재 주피터 노트북 파일의 경로
    - ls 현재 디렉토리의 파일 리스트
    - whos 현재 선언된 변수를 출력
    - reset 현재 선언된 변수를 모두 삭제 
```

```python
%pwd
```

```python
### Shell command
- 주피터 노트북을 실행 쉘 환경의 명령을 사용
- 명령어 앞에 !를 붙여서 실행
- 주요 명령어
    - ls, cat, echo......
```
