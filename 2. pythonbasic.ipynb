3. 식별자

- 식별자 규칙

- 컬렉션 데이터 타입 : list, tuple, dict

# 컬렉션 데이터 타입 : llist, tuple, dict

- list [] : 순서가 있는 수정이 가능한 데이터 타입
- tuple () : 순서가 있고 수정이 불가능한 데이터 타입
- dict {} : 순서가 없고 키:값 으로 구성되어 있는 데이터 타입

```python
# list
ls = [1, 2, 3, "Four", [5, 6], True, 1.2]
type(ls), ls
```

```python
# offset index 사용이 가능
ls[3], ls[1:3], ls[::-1]
```

```python
# list 함수
ls = [1, 5, 2, 4]
# append : 가장 뒤에 값을 추가
ls.append(3)
ls
```

```python
#sort : 오름차순으로 정렬
ls.sort()
ls
```

```python
# 내림차순
ls[::-1]
```

```python
# pop : 가장 마지막 데이터를 출력하고 출력한 데이터를 삭제
# ctrl + enter : 현재 셀을 계속 실행
num = ls.pop()
num, ls
```

```python
# 리스트의 복사
ls1 = [1, 2, 3]
ls2 = ls1 # 얕은복사 : 주소값복사
ls1, ls2
```

```python
ls1[2] = 5
ls1, ls2
```

```python
ls3 = ls1.copy()
ls1, ls2, ls3
```

```python
ls1[2] = 10
ls1, ls2, ls3
```

```python
# tuple
# 튜플은 리스트보다 같은 데이터를 가졌을때 공간을 적게 사용

tp1 = 1, 2, 3
tp2 = (4, 5, 6)
type(tp1), type(tp2), tp1, tp2
```

```python
a, b = 1, 2
a, b
```

```python
#offset index 사용
tp1[1], tp1[::-1]
```

```python
# 리스트와 튜플의 저장공간 차이 비교
import sys

ls = [1, 2, 3]
tp = (1, 2, 3)

print(sys.getsizeof(ls), sys.getsizeof(tp))
```

```python
#dict {} : 키는 정수, 문자열 데이터 타입만 사용이 가능
#순서가없고 {키:값}으로 구성되어 있는 데이터 타입

dic = {
    1:"one",
    "two":2,
    "three":[1,2,3]
}
type(dic), dic
```

```python
dic[1], dic["three"]
```

```python
dic["two"] = 123
dic
```

```python
# 아래의 데이터를 list 와 dict으로 선언
# 도시 : seoul, busan, daegu
# 인구 : 9,700,000, 3,400,000, 2,400,000

#리스트
city = ["seoul", "busan", "daegu"]
population = [9700000, 3400000, 2400000]

#딕셔너리
data = {
    "seoul":9700000,
    "busan":3400000,
    "daegu":2400000
}

city, population, data
```
