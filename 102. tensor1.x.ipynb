```python
import tensorflow as tf
```

```python
hello = tf.constant("Hello World")
sess = tf.Session()
sess.run(hello)
```

```python
# constant 상수
a = tf.constant(1)
b = tf.constant(2)
c = tf.add(a, b)
sess = tf.Session()
sess.run(c)
```

```python
# Variable 변수
a = tf.Variable(5)
b = tf.Variable(3)
c = tf.multiply(a, b)
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
sess.run(c)
```

```python
# placeholder 학습 데이터를 포함하는 변수

input = [1, 2, 3, 4, 5]

x = tf.placeholder(dtype=tf.float32)
y = x + 5

sess = tf.Session()
sess.run(y, feed_dict={x:input})
```

```python
mathscore = [85, 99, 84, 97, 92]
englishscore = [59, 80, 84, 68, 77]

a = tf.placeholder(dtype=tf.float32)
b = tf.placeholder(dtype=tf.float32)
y = (a + b) / 2

sess = tf.Session()
sess.run(y, feed_dict={a:mathscore, b:englishscore})
```

```python
# 주요 함수

# tf.add : 덧셈
# tf.subtract : 뺄셈
# tf.multiply : 곱셈
# tf.truediv : 나눗셈의 몫
# tf.mod : 나눗셈의 나머지
# tf.abs : 절대값

# tf.negative : 음수
# tf.sign : 부호
# tf.square : 제곱
# tf.pow : 거듭제곱
# tf.maximum : 큰값
# tf.minimum : 작은값
# tf.exp : 지수값
# tf.log : 로그값 
# tf.sin : 사인값
# tf.cos : 코사인값
```

```python
a = tf.constant(17)
b = tf.constant(5)

c = tf.add(a, b)

sess = tf.Session()
sess.run(c)
```
