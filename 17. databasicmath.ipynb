```python
A = frozenset([])
B = frozenset(['H'])
C = frozenset(['T'])
D = frozenset(['H', 'T'])

P = {A: 0, B: 0.4, C: 0.6, D:1}
P
```

# 확률의 의미

- 빈도주의적 관점 : 선택된 표본이 사건의 원소가 될 경향
- 베이지안 관점 : 선택된 표본이 특정한 사건에 속한다는 가설 명제 혹은 주장의 신뢰도라고 본다

# 확률분포

- 사건이라는 표본의 집합에 대해 숫자를 할당하는 함수

# 단순사건과 확률질량함수

- 단순사건에 대한 확률만 정의하는 함수를 확률질량함수라고 한다

- P({1}) = 0.2   원소가 하나뿐인 사건
- P(1) = 0.2     확률질량함수

- P({1, 2}) = 0.3  <-- 가능
- P(1, 2)   <-- 틀린식

# 표본의 수가 무한한 경우

- 표본수가 무한하고 모든 표본에 대해 표본 하나만을 가진 사건의 확률이 동일하다면 표본 하나에 대한 사건의 확률은 언제나 0이다.
- 표본수가 무한하면 사건에 대해 직접 확률을 할당할수밖에 없다는 것 을 알수있다.

# 구간

- P(a, b) = 수식

# 누적분포함수

- P(a, b) = F(b) - F(a)

- 음의 무한대에 대한 누적분포함수값은 0이다.
- 양의 무한대에 대한 누적분포함수값은 1이다.
- 입력이 크면 누적분포함수값은 커진다.

- 누적 분포함수 --> 미분 --> 확률 밀도함수(도함수)

# 결합확률과 조건부확률

- 결합확률 : A 와 B가 동시에 발생할 확률 
- 조건부확률 : B가 사실일 경우 사건 A에 대한 확률을 B에 대한 서건 A의 조건부확률 이라함

# 독립

- 새로운 B가 주어지든 아니든 A가 범인일 확률이 변함이 없다 

# 사슬법칙

- 조건부확률과 결합확률의 관계를 확장하면 복수의 사건에 대한 조건부 확률을 사슬법칙이라 한다

# 대칭분포

- 분포가 표본평균을 기준으로 대칭인 대칭분포이면 표본중앙값은표본평균과 같다.
- 분포가 대칭분포이면서 하나의 최고값만을 가지는 당봉분포이면 표본최빈값은 표본평균과 같다.
- 대칭분포를 비대칭으로 만드는 데이터가 더해지면표본평균이 가장 크게 영향을 받고 표본최빈랎이 가장 적게 영향을 받는다.
